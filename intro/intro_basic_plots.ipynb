````python
import matplotlib.pyplot as plt # Plotting library
import seaborn as sns # Plotting library
import pandas as pd # Main data cleaning & manipulation library
import numpy as np # Statistics library
````

First, let's load some data to visualize. We'll start with the tips dataset from seaborn. Description: One waiter recorded information about each tip he received over a period of a few months working in one restaurant. 

He collected several variables:
- tip in dollars
- bill in dollars
- sex of the bill payer
- whether there were smokers in the party
- day of the week
- time of day
- size of the party

In all he recorded 244 tips. The data was reported in a collection of case studies for business statistics (Bryant & Smith 1995).

````python
# Load some sample data
tips = sns.load_dataset("tips")
tips.head()
````

## Basic Visualization

Seaborn has some great eample datasets to use for visualizations. See the whole list here: https://github.com/mwaskom/seaborn-data

The normal syntax for a seaborn plot is ```sns.[PLOT-TYPE](...``` See a list of plot types here: https://seaborn.pydata.org/api.html#relational-api

Data visualizations are meant to visuzlie the answers to questions. What question does the bar chart below answer?

````python
# Bar Plot
sns.barplot(x="day", y="total_bill", data=tips) # Create the bar plot
plt.title("Average Total Bill per Day") # Add a title
plt.show()
````

What type of chart would you use to visualize the **distribution** of tips? Make the chart.

````python
# Scatter Plot
sns.histplot(data = tips, x = "tip") # Create the scatter plot
# plt.title("Total Bill vs Tip") # Add a title
plt.show()
````

The distribution of tips is right-skewed, meaning most tips are on the lower end with fewer high tips. Does this suggest that customers generally tip well at this restaurant, or could there be another explanation for this pattern?

What question does the scatter chart below answer?

````python
# Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips) # Create the scatter plot
plt.title("Total Bill vs Tip") # Add a title
plt.show()
````

## Customizing Visualizations

Let's customize the scatterplot above a bit:
- Color the points by the day of the week
- Style the points according to lunch/dinner
- Add x and y axis labels
- Create a legend

Check out the documentation for scatterplots here: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

````python
plt.figure(figsize=(8, 6)) # 8 wide, 6 high
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips) # hue = color of a point, style = shape of a point
plt.title("Tips by Day and Time")
plt.xlabel("Total Bill ($)") # X-axis label
plt.ylabel("Tip ($)")# Y-axis label
plt.legend(title="Day/Time") # Create a legend
plt.show()
````

````python
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Distribution of Tips by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount ($)")
plt.show()
````

````python
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 6))
sns.barplot(x="class", y="survived", hue="sex", data=titanic)
plt.title("Titanic Survival Rates by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
````

````python
“How do different penguin species compare across physical measurements? Are any features especially useful for distinguishing between species?”

Discussion Questions:

"Which pair of features best separates the different species, and how can you tell?"
Encourages students to look for clear clusters or separations in the scatter plots.
"If you were to build a model to classify penguin species, which features would you prioritize based on this plot?"
````

When visualizations have a lot going on, it's easy to glance at them quickly and move on, but let's dig into this one!

- Similar to a heatmap, but visualizing three variables
- The objective of the plot is to visualize how the relationhips of differnt peguin traits (Bill Length (mm), Bill Depth (mm), Flipper Length (mm), Body Mass (g)) are similar/differnt among three differnt species (Adelie, Chinstrap, Gentoo)
- Each grid cell shows that relationship between two different variables within 3 specices. For example, the rightmost top cell show the relationship between Bill Length (mm) and Body Mass (g)
- The diagonal cells with the curves show the relationships between the same variables? Why are the curves skewed and not symmetrical within a species?

````python
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins, hue="species", diag_kind="kde")
plt.suptitle("Pairwise Relationships in Penguin Species", y=1.02)
plt.show()
````
