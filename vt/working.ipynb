```python
from scipy.stats import zscore
import pandas as pd
```

```python
df = pd.read_csv("./Airbnb_Open_Data.csv")
df.head()
```

```python
zscore(df["price"])
```

```python
df.dtypes
```

```python
df["price"] = df["price"].str.replace("$", "")
df["price"] = df["price"].str.replace(",", "")
df["price"] = df["price"].astype(float)
```

```python
zscore(df["price"], nan_policy = "omit")
```

```python
df["minimum nights"] = df["minimum nights"].astype(float)
df["minimum_nights_zscore"] = zscore(df["minimum nights"], nan_policy = "omit")
df.head()
```

```python
df[df["minimum_nights_zscore"] > 3]
```
